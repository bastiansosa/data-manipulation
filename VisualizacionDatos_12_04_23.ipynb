```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.pairplot(iris, hue="species")
```

```python
iris.corr()
```

```python
sns.heatmap(data=iris.corr(),cmap="viridis", annot=True)
```

```python
pd.plotting.parallel_coordinates(iris,'species')
```

```python
df = pd.read_csv("Life Expectancy Data.csv")
df.info()
```

```python
colors = {"Developed":"red","Developing":"blue"}
sns.scatterplot(data=df,x="Year", y="Life expectancy ",hue="Status",palette=colors) #hue_order->para indicar el orden 
```

```python
sns.countplot(data=df, x="Status")
```

```python
sns.countplot(data=df,x='Year',hue="Status")
plt.xticks(rotation=90)
```

```python
sns.relplot(data=df,x="Year", y="Life expectancy ",kind="scatter", col="Status")
```

```python
sns.relplot(data=df,x="infant deaths", y="Life expectancy ",kind="scatter", col="Status",row="Year")
```

```python
sns.relplot(data=df,x="Year", y="Life expectancy ",kind="scatter", col="Status", size="Schooling",hue="Schooling")
```

```python
df_filter = df[   (df.Country=="Chile") | (df.Country=="Uruguay")   ]

df_filter.columns.values
```

```python
sns.relplot(data=df_filter, x="Year", y="Life expectancy ", kind="line", hue="Country") 
```
